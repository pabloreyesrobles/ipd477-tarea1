```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
```

```python
f0 = 40
t0 = 1 / f0
fs = 1000
ts = 1 / fs

t = np.arange(0, t0, ts)
x = 2 * np.sin(2 * np.pi * f0 * t)
plt.plot(t, x)

print('enería en un ciclo: {:.4f}'.format(np.trapz(x ** 2, t)))
print('energía r/ tasa de muestreo: {:.4f}'.format(ts * np.sum(x ** 2)))
```

```python
t = np.arange(0, 10 * t0, ts)
x = 2 * np.sin(2 * np.pi * f0 * t)
plt.plot(t, x)

print('enería en 10 ciclos: {:.4f}'.format(np.trapz(x ** 2, t)))
print('energía r/ tasa de muestreo: {:.4f}'.format(ts * np.sum(x ** 2)))
print('potencia media de la señal: {:.4f}'.format(np.trapz(x ** 2, t) / (t[-1] - t[0])))
print('potencia media de la señal periódica: {:.4f}'.format(np.sum(x ** 2) / x.shape[0]))
```

```python
N = 600
T = 1.0 / 800.0

t = np.linspace(0.0, N * T, N, endpoint=False)
x = np.sin(50.0 * 2.0 * np.pi * t) + 0.5 * np.sin(80.0 * 2.0 * np.pi * t)

xf = fft(x)[:x.shape[0] // 2]
ff = fftfreq(x.shape[0], ts)[:x.shape[0] // 2]
plt.plot(ff, 2 / N * np.abs(xf))
```
